```python
import zipfile

zip_path = "Bike Helmet Detection.v2-more-preprocessing-augmentation.yolov8.zip"

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall("helmet-dataset")

print("Unzipped to folder: helmet-dataset/")
```

```python
from ultralytics import YOLO

# Load a YOLOv8 pre-trained model (you can pick yolov8n, yolov8s, etc)
model = YOLO('yolov8n.pt')

# Train the model on your dataset
model.train(data='helmet-dataset/data.yaml', epochs=30, imgsz=640)
```

```python
import torch
print("CUDA available:", torch.cuda.is_available())
```

```python
metrics = model.val()
print(metrics)
```

```python
# Run inference on an image
results = model("https://c8.alamy.com/comp/KR8624/family-on-motorbike-with-only-the-driver-wearing-a-helmet-on-road-KR8624.jpg")

# Visualize results
results[0].show()
```

```python
model.save("helmet-detection-final.pt")
```
